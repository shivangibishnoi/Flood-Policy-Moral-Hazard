```python
import altair as alt
import pandas as pd
```

```python
# import dataset
df = pd.read_csv("world-data.csv")
df
```

```python
# create multiple line chart
line = alt.Chart(df).mark_line().encode(
    x='Year:O',
    y='Value:Q',
    color='Category:N'
).properties(
    width=600,
    height=400
)

```
